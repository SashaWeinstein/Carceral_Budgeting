Next question: how much of the GIC's total budget goes to insurance related expenses?

```python
GIC = pd.DataFrame(client.get("pegc-naaa",
                              where="Department = 'GROUP INSURANCE COMMISSION (GIC)' \
                              AND budget_fiscal_year >= 2016 AND budget_fiscal_year <= 2019",
                              limit=999999))
```

```python
GIC["amount"] = GIC["amount"].astype(float)
GIC.groupby("object_class").sum()["amount"]
```

```python
GIC.groupby("object_code").sum()["amount"].sort_values(ascending=False)
```

Ok cool glad to dobule check. It spends around 4 million on payroll and around around 2.5 billion per year on insurance related expenses. All this money should be counted towards the fringe benefits as the cost of administering it is part of providing the benefit.
