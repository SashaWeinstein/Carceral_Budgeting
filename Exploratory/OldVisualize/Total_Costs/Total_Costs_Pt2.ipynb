Next up I want to look specifically at how much pensions/hidden payroll costs effect each agency. Visualize this with pre-pension number as dotted line and post-pension/hidden payroll number as solid line

```python
pre_pension_costs = Total_Costs_1Ver.get_total_cost(list(range(2016, 2020)))
```

```python
temp = ["Without Hidden Costs"] * pre_pension_costs.shape[0]
pre_pension_costs["With Hidden Costs"] = temp
```

```python
temp = ["With Hidden Costs"] * total_cost.shape[0]
total_cost["With Hidden Costs"] = temp
```

```python
total_cost.reset_index(inplace=True)
```

```python
big = total_cost.append(pre_pension_costs.reset_index())
big
```

```python
budgets_for_melt = big[(big["index"].str.contains("Budget") )&
                       (big["index"].str.contains("Total") == False)]\
                      .loc[:,["index"] +list(range(2016,2020))+["With Hidden Costs"]]
```

```python
budgets_for_melt
```

```python
big_melted = budgets_for_melt.melt(id_vars=["index", "With Hidden Costs"])
```

```python
big_melted.rename(columns={"variable":"year"},inplace=True)
```

```python
big_melted["value_in_millions"] = big_melted["value"]/10**6
palette = sns.color_palette("Paired", melted_budget["index"].nunique())
fig, ax = plt.subplots(1, 1)
plt.ticklabel_format(style='plain', axis='y')
p = sns.lineplot(x="year", y="value_in_millions", style = "With Hidden Costs",
                 hue="index", palette=palette, sort=False, data=big_melted)
p.set_title("Budgets Over Time with and without Hidden Costs", fontsize=24)
ax.legend(frameon=False, loc="center right", fontsize=16)
plt.xticks(list(range(2016,2020)))
p.set_ylabel("Budgeted Dollars in Millions", fontsize=20)
p.set_xlabel("")
p.tick_params(labelsize=20)
```

Ok great, do this without boston PD so we can see the other agencies in detail

```python
palette = sns.color_palette("Paired", big_melted_no_BostonPD["index"].nunique())
fig, ax = plt.subplots(1, 1)
plt.ticklabel_format(style='plain', axis='y')
p = sns.lineplot(x="year", y="value_in_millions", style = "With Hidden Costs",
                 hue="index", palette=palette, sort=False, data=big_melted_no_BostonPD)
p.set_title("Budgets Over Time with and without Hidden Costs", fontsize=24)
ax.legend(frameon=False, loc="center right", fontsize=16)
plt.xticks(list(range(2016,2020)))
p.set_ylabel("Budgeted Dollars in Millions", fontsize=20)
p.set_xlabel("")
p.tick_params(labelsize=20)
```
