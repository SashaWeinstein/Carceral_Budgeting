# Define some functions

```python
def old_method(df):
    print("number of payroll expenditures with old methodolgy:",
      df[df["object_code"].str[1]=="A"].shape[0])
    print("total money spent on payroll with old methodolgy",
         df[df["object_code"].str[1]=="A"]["amount"].sum())
def new_method(df):
    print("number of payroll expenditures with new methodolgy:",
      df[df["vendor"].str.contains("PAYROLL")].shape[0])
    print("total money spent on payroll with old methodolgy",
        df[df["vendor"].str.contains("(?i)PAYROLL")]["amount"].sum())
```
