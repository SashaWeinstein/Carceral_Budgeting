```python
HRD_official_name = "HUMAN RESOURCES DIVISION (HRD)"
HRD = pd.DataFrame(client.get("pegc-naaa",
                             where = "Department = '" + HRD_official_name + "' AND budget_fiscal_year = 2019",
                  limit =999999))
```
