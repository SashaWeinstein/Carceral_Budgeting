```python
pension_payouts = pd.DataFrame(client.get("pni4-392n", where = "year >=2016 AND year <= 2019", limit=999999))
```

```python
cthru_pensions = pension_payouts
cthru_pensions.loc[:,"annual_amount"] = cthru_pensions.loc[:,"annual_amount"].astype(float)
cthru_pensions = cthru_pensions[cthru_pensions["retirement_system"] == "MSERS"]
cthru_pensions.rename(columns={"department_last_worked_in": "department",
                               "title_at_retirement":"title",
                               "annual_amount":"amount"}, inplace=True)
cthru_pensions["department"].fillna("None", inplace=True)
cthru_pensions["year"] = cthru_pensions["year"].astype(int)
cthru_pensions["title"].fillna("None", inplace=True)
```

## Look at Mass Dept of Transportation

```python
mDOT_pensions= cthru_pensions[cthru_pensions["department"]=="Mass Dept Of Transportation"]
amount_by_title = mDOT_pensions.groupby("title").sum()["amount"].sort_values(ascending=False)
```
