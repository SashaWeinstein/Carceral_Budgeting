## Case Filings

```python
from Efficiency_Metrics import total_estimated_incidents_including_unreported_2016, \
total_reported_criminal_incidents_2016, total_case_filings_2016, \
case_filings_2016_leading_to_conviction
```

## Create case filings series

```python
unreported_2016 = total_estimated_incidents_including_unreported_2016 - total_reported_criminal_incidents_2016
unreported_2016
```

```python
reported_no_case_filing_2016 = total_reported_criminal_incidents_2016 - total_case_filings_2016
reported_no_case_filing_2016
```

Then look at how many case filings don't result in convictions

```python
case_filings_no_conviction_2016 = total_case_filings_2016- case_filings_2016_leading_to_conviction
case_filings_no_conviction_2016
```

Make sure these all sum back up to total reported 

```python
assert case_filings_2016_leading_to_conviction+ case_filings_no_conviction_2016+\
reported_no_case_filing_2016 ==total_reported_criminal_incidents_2016
```

```python
assert case_filings_2016_leading_to_conviction + \
case_filings_no_conviction_2016 +\
reported_no_case_filing_2016 +\
unreported_2016 == total_estimated_incidents_including_unreported_2016
```

```python
case_filing_series = pd.Series(index=["Unreported Incidents (Estimated)",
    "Reported Incidents that don't \n lead to Case Filing",
                                      "Incidents that lead to Case Filing \n but no Conviction",
                                      "Incidents that ultimately \n lead to Conviction"],
                               data=[unreported_2016, reported_no_case_filing_2016, 
                                     case_filings_no_conviction_2016,
                                     case_filings_2016_leading_to_conviction])
```

```python
case_filing_series.iloc[1:]
```

## Case filing analysis out of reported

```python
cr = [1, 3,4]
reported_to_case_filings = pie_plot(case_filing_series.iloc[1:], "Reported incidents leading to Convictions \n2016", True,
                                  cr)
reported_to_case_filings.savefig("CaseFlow_Pies/Case_Filing_rate_Reported.png")
```

## Case filings analysis out of unreported

```python
cr = [0,1, 3,4]
unreported_to_case_filings = pie_plot(case_filing_series, "Unreported Incidents leading to Convictions \n2016", True, cr)
unreported_to_case_filings.savefig("CaseFlow_Pies/Case_Filing_rate_Unreported.png")
```
